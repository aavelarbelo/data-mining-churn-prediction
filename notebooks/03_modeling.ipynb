```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)

df = pd.read_csv("../data/processed/telco_churn_clean.csv")
print(df.shape)
df.head()
```

```python
print("Ausentes no total (deve ser 0):", df.isna().sum().sum())
print()
print("Alvo Churn (deve ter só No e Yes):")
print(df["Churn"].value_counts())
```

```python
# 1) remover o identificador (não é característica de comportamento)
df_model = df.drop(columns=["customerID"])

# 2) separar características (X) e alvo (y)
X = df_model.drop(columns=["Churn"])
y = df_model["Churn"]

# 3) one-hot encoding: cada categoria vira uma coluna 0/1
X = pd.get_dummies(X)

print("Formato de X depois do encoding:", X.shape)
print("Formato de y:", y.shape)
X.head()
```

```python
sorted(X.columns.tolist())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,      # 20% guardado para o exame final
    stratify=y,          # manter a proporção 73/27 nas duas fatias
    random_state=42      # semente fixa: divisão sempre igual (reprodutibilidade)
)

print("Treino:", X_train.shape, "| Teste:", X_test.shape)
print()
print("Proporção de churn no treino:")
print(y_train.value_counts(normalize=True).round(3))
print()
print("Proporção de churn no teste:")
print(y_test.value_counts(normalize=True).round(3))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# 1) criar o modelo (k=5: olhar para os 5 vizinhos mais parecidos)
knn = KNeighborsClassifier(n_neighbors=5)

# 2) treinar: o modelo "estuda" com os dados de treino
knn.fit(X_train, y_train)

# 3) exame: prever nos dados de teste (que ele nunca viu)
previsoes = knn.predict(X_test)

# 4) nota do exame: % de acertos
print("Accuracy do KNN:", round(accuracy_score(y_test, previsoes), 3))
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print("Matriz de confusão (linhas = realidade, colunas = previsão):")
print(confusion_matrix(y_test, previsoes, labels=["No", "Yes"]))
print()
print(classification_report(y_test, previsoes))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import precision_score, recall_score, f1_score

modelos = {
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "SVM": SVC(random_state=42),
    "Naive Bayes": GaussianNB(),
}

resultados = []
for nome, modelo in modelos.items():
    modelo.fit(X_train, y_train)
    pred = modelo.predict(X_test)
    resultados.append({
        "Modelo": nome,
        "Accuracy": round(accuracy_score(y_test, pred), 3),
        "Precision": round(precision_score(y_test, pred, pos_label="Yes"), 3),
        "Recall": round(recall_score(y_test, pred, pos_label="Yes"), 3),
        "F1": round(f1_score(y_test, pred, pos_label="Yes"), 3),
    })

tabela = pd.DataFrame(resultados).set_index("Modelo")
print(tabela)
```

```python
from sklearn.preprocessing import StandardScaler

# escalar: todas as colunas passam a ter média 0 e desvio 1 (ninguém grita mais alto)
scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)   # aprende a escala NO TREINO...
X_test_sc = scaler.transform(X_test)         # ...e aplica a mesma ao teste (sem batota)

resultados_sc = []
for nome, modelo in modelos.items():
    modelo.fit(X_train_sc, y_train)
    pred = modelo.predict(X_test_sc)
    resultados_sc.append({
        "Modelo": nome,
        "Accuracy": round(accuracy_score(y_test, pred), 3),
        "Precision": round(precision_score(y_test, pred, pos_label="Yes", zero_division=0), 3),
        "Recall": round(recall_score(y_test, pred, pos_label="Yes"), 3),
        "F1": round(f1_score(y_test, pred, pos_label="Yes"), 3),
    })

tabela_sc = pd.DataFrame(resultados_sc).set_index("Modelo")
print("COM escalonamento:")
print(tabela_sc)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer

f1_yes = make_scorer(f1_score, pos_label="Yes")

print("Validação cruzada (5 folds, métrica = F1 da classe Yes):")
for nome, modelo in modelos.items():
    scores = cross_val_score(modelo, X_train_sc, y_train, cv=5, scoring=f1_yes)
    print(f"{nome:15s}  F1 médio: {scores.mean():.3f}  (± {scores.std():.3f})")
```

```python
import os
os.makedirs("../reports", exist_ok=True)
tabela_sc.to_csv("../reports/model_comparison.csv")
print("Tabela guardada em reports/model_comparison.csv")
print(tabela_sc)
```
